# Create RDD from Local File

```python
docker_file = context.textFile("./work/Dockerfile")
```

```python
## RDD1
docker_file
```

```python
docker_file.take(10)
```

```python
def loop(txt):
    # print(txt)
    return txt.split(" ")

# Tranformation (2nd RDD)
new_rdd = docker_file.map(loop)
```

```python
# Action
new_rdd.take(100)
```

```python
new_rdd.take(100)
```

```python
new_rdd.take(100)
```

```python
rdd3 = new_rdd.flatMap(lambda x: x)
```

```python
rdd3.take(10)
```

```python
def reduce(prev, new):
    return prev+new

word_count = rdd3.map(lambda key: (key, 1)).reduceByKey(reduce)
```

```python
word_count.collect()
```

```python
word_count.collect()
```

```python
text = "Hello world"*10000000
txts = text.split(" ")
txts[:10]
```

```python
text = "Hello world"
rdd = context.parallelize([text])
```

```python
rdd.collect()
```

```python
def multiply_rec(txt):
    return txt*100
    
new_rdd = rdd.map(multiply_rec).map(multiply_rec)
```

```python
splitted = new_rdd.map(lambda txt: txt.split(" "))
```

```python
flatted = splitted.flatMap(lambda x: x)
flatted
```

```python
flatted.count()
```

```python
flatted.take(10)
```

# WordCount

```python
rdd = context.textFile("hdfs://namenode:9000/data/films")
rdd.take(1) # Risk -> it will read the your data, if your data 1PB
```

```python
## High Perfomance

wordcount_rdd = (rdd.map(lambda x: x.split(","))
    .flatMap(lambda x: x)
    .map(lambda k: (k, 1)) # MapReduce Style
    .reduceByKey(lambda total, newcount: total+newcount)
)
```

```python
wordcount_rdd.take(10)
```

```python
## Low Performance
```

```python
rdd.map(lambda line: line.split(","))\
        .flatMap(lambda word: word)\
       .groupBy(lambda word: word) \
        .mapValues(lambda groupedWord: len(groupedWord)) \
        .take(10)
```
